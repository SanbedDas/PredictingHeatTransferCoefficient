## Defining the features and the target

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

## Splitting the data into test and training sets

```python
from sklearn.model_selection import train_test_split as tts 
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=10)
```

## Scaling the Features

```python
from sklearn.preprocessing import StandardScaler as ss
sc=ss()
x_train[:,:]=sc.fit_transform(x_train[:,:])
x_test[:,:]=sc.transform(x_test[:,:])
y_train=sc.fit_transform(y_train[:].reshape(-1,1))
y_test=sc.fit_transform(y_test[:].reshape(-1,1))
```

## Training the SVR model with the best Hyperparameters

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
regressor = SVR()
# Defining the hyperparameters to tune and the range of values
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 10], 'kernel': ['linear', 'rbf', 'poly', 'sigmoid']}
grid = GridSearchCV(estimator=regressor, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, verbose=1, n_jobs=-1)
# Fitting the grid search object to the data
grid.fit(x_train, y_train.ravel())
```

```python
# Printing the best hyperparameters and the best score
print("Best hyperparameters: ", grid.best_params_)
print("Best score: ", -grid.best_score_)
```

```python
y_pred = grid.predict(x_test)
np.set_printoptions(precision=2)
y_pred=sc.inverse_transform(y_pred.reshape(-1,1))
y_test=sc.inverse_transform(y_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Calculating the Evaluating metrics

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

## Data Visualisation

```python
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('True vs. Predicted Values')
plt.show()
```

## Residuals Plot

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--', lw=2)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

## Histogram of Residuals

```python
plt.hist(residuals, bins=50)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.show()
```
